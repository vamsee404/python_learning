```python
'''
Problem Statement

Build a Login System

User has Maximum 5 attepts to enter the correct password
The Correct password is "RajCloud1234"
If the user enters an empty password, skip that attemp( Do not count it )
If the user enters correct password show success message and stop the loop
If the user enters a erong password count the attempt
If 3 wrong password happen, Block user and stop the loop
'''
count = 0

for i in range(6):
    user_input = input('Please enter the password: ')

    if user_input == '':
        continue

    if user_input == 'RajCloud1234':
        print('You are successfully logged in')
        break
    else:
        count += 1

    if count == 3:
        print('You have entered the wrong password 3 times. User is blocked.')
        break

```
